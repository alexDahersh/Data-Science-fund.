I'm going to use 2022 and 23 for my prior and 2024 for my data. The difference between these two years and 2024 can clearly be seen above. The seperation here is when the heat dome covered Austin and 2024 will be the data since it's more recent.

I'm also going to roughly half the prior since I don't think the past two years should be weighted too highly. If I don't halve the prior then these past years will overpower this year in the posterior. The prior is supposed to be just a guess so I don't want it to overpower the data.


```python
!pip install preliz # Colab doesn't have preliz installed so this is necessary

import preliz as pz # importing the rest of the libraries
import arviz as az
import pymc as pm
```

```python
prior_Beta = ((total22+total23)-(totalAbove22+totalAbove23))/2 # Alpha which is half the combined data from 2022 and 2023
prior_Alpha = (totalAbove22+totalAbove23)/2 # Beta which is half the combined data from 2022 and 2023

weather_Prior = pz.Beta(alpha=prior_Alpha,beta=prior_Beta) # Beta distribution from combined data of 2022 and 2023 that was halved.
weather_Prior.plot_pdf() # Plotting prior
```

```python
# A visual representation of the prior ^. The mean is about 65%. I didn't intend for the alpha and beta to not be whole numbers,
# but I'm keeping it since it seems to work
```

```python
weather_Likelihood = pz.Binomial(n = total24, p = totalAbove24 / total24,) # Beta distribution that with number of summer days in 2024 as its n and the percent of days above 100 in 2024 for its probabiliyu
weather_Likelihood.plot_pdf()
```

Creating a posterior:

The alpha from the prior and the times it happened from the likelihood are added for a new prior.

The beta from the prior is added to the possibles times it (days above 100 degrees) could have happened from the data.

```python
newAlpha = prior_Alpha + totalAbove24 # First part is prior plus data from 2024
newBeta = prior_Beta + (total24 - totalAbove24) # First part is from prior, second is from 2024

posterior = pz.Beta(alpha = newAlpha, beta = newBeta)
posterior.plot_pdf()
posterior.mean()
```

The posterior predicts that 46% of the days in summer next year will be above 100 degrees. The degree of certainty is shown in the graph above.

I'm now going to do this by drawing samples from the posterior using pymc because why not

```python
weather_Above100 = np.repeat(1,totalAbove24) # Numpy array of 1s for all the days above 100
weather_Below100 = np.repeat(0,total24 - totalAbove24) # Numpy array of 0s for all the days below 100

weather_Data = np.concatenate([weather_Above100, weather_Below100]) # Combining the days above and below 100 for the complete data
```

```python
with pm.Model() as austin_weather_model: # Sampling from posterior
  theta = pm.Beta('' , alpha = float(prior_Alpha) , beta = float(prior_Beta)) # The prior

  y = pm.Bernoulli('y', p = theta, observed = weather_Data) # The likelihood

  idata_Weather = pm.sample(10000) # 10 thousand because more is better

az.plot_posterior(idata_Weather) # Plotting samples
```

```python
print(idata_Weather.posterior.mean()) # Printing the mean of the samples
```

```python
pm.sample_posterior_predictive(idata_Weather, model=austin_weather_model, extend_inferencedata=True) # Generating future prediction
```

```python
idata_Weather.posterior_predictive.mean() # Mean of future predictions
```

```python
stacked_Weather = az.extract(idata_Weather, group="posterior_predictive") # Not entirely sure what this does, but it combines the data
stacked_Weather.sel(sample=0).sel(draw=0).y.to_series().values # Prints predictions for which days of summer in 2025 will be above 100 degrees
```

So, the posterior. I think it's pretty subpar for predictions. The mean of the likelihood was roughly .25 and the mean of the prior was around .65. Neither of these numbers are even on the the graph, let alone in the 94% range. Even in 2022, the year that was the most average, about 56% of the days were above 100 degrees, still not in the 94% range. The reason for this was because I had the prior and the likelihood have about the same weight. The posterior is kind of like the average of the two. But, based on the heat dome and the science behind this, the average of the two doesn't make much sense (at least during the years I looked at).

For the true probability, I think it's also pretty subpar. The way its made almost assumes Austin will have a heat dome half the time.

Also, the temperatures between subsequent days isn't random. The predictions the model did assumed this and therefore likely gave bad results. Hot days are often next to hot days and cold days are often next to cold days.
